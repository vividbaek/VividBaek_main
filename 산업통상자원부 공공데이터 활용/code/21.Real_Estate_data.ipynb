### tong 데이터에서 잘못된 값 삭제 및 올바르게 mapping

```python
location_mapping = {
    '경기': '경기도',
    '경기도': '경기도',
    '충남': '충청남도',
    '충청남도': '충청남도',
    '충북': '충청북도',
    '충청북도': '충청북도',
    '인천': '인천광역시',
    '인천광역시': '인천광역시',
    '전북': '전라북도',
    '전라북도': '전라북도',
    '경북': '경상북도',
    '경상북도': '경상북도',
    '경남': '경상남도',
    '경상남도': '경상남도',
    '서울': '서울특별시',
    '강원': '강원도',
    '강원도': '강원도',
    '대전': '대전광역시',
    '부산': '부산광역시',
    '세종특별자치시': '세종특별자치시',
    '세종시': '세종특별자치시',
    '울산광역시': '울산광역시',
    '울산': '울산광역시',
    '전남': '전라남도',
    '전라남도': '전라남도',
    '제주': '제주특별자치도',
    '제주특별자치도': '제주특별자치도',
    '소화성시': '경기도',
    '반월': '경기도',
    '시흥시': '경기도',
    '화성시': '경기도',
    '안산시': '경기도',
    '-경기': '경기도',
}

# NaN 값을 빈 문자열로 변환
tong['매물위치(1)'] = tong['매물위치(1)'].fillna('')

# 매물위치(1) 값을 올바르게 매핑
tong['매물위치(1)'] = tong['매물위치(1)'].map(location_mapping)

# NaN 값을 다시 빈 문자열로 변환
tong['매물위치(1)'] = tong['매물위치(1)'].fillna('')

# 잘못된 값 삭제
tong = tong[~tong['매물위치(1)'].str.startswith('S|-')]
tong = tong[tong['매물위치(1)'] != '']
```
